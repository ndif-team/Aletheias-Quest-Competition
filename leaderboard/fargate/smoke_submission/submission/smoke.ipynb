```python
import os, pandas as pd
from datasets import load_dataset
# Loads the BAKED eval inputs offline (HF_DATASETS_OFFLINE=1 + baked arrow cache),
# proving the sandboxed offline-dataset path works on Fargate. No NDIF needed.
ds = load_dataset(os.environ["DATASET_NAME"], split="test")
df = ds.to_pandas()
idx = df["index"].tolist() if "index" in df.columns else list(range(len(df)))
n = len(idx)
# Trivial deterministic 'detector' just to exercise scoring end to end.
score = [(i % 5) / 4 for i in range(n)]
deceptive = [s >= 0.5 for s in score]
pd.DataFrame({"index": idx, "deceptive": deceptive, "score": score}).to_csv(
    "submission.csv", index=False)
print("wrote", n, "rows")
```
