```python
import pandas as pd
# Deterministic, offline submission used to test the runner without NDIF/HF.
# labels.csv = [1, 1, 0, 0, 1]; deceptive is perfect, score separates perfectly.
index = [0, 1, 2, 3, 4]
deceptive = [True, True, False, False, True]
score = [0.9, 0.8, 0.2, 0.1, 0.6]
pd.DataFrame({"index": index, "deceptive": deceptive, "score": score}).to_csv(
    "submission.csv", index=False)
print("wrote", len(index), "rows")
```
